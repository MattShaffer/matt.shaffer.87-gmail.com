```python
import pandas as pd

file_to_load = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)
```

```python
Player_info=len(purchase_data["SN"].unique())
Totalspurchase_data = pd.DataFrame({"Total Players": [Player_info]})
Totalspurchase_data.head()
```

```python
item_ID=len(purchase_data["Item ID"].unique())
av_price=purchase_data["Price"].mean()
num_purchase=len(purchase_data["Purchase ID"].unique())
total_rev=purchase_data["Price"].sum()
av_price='${:,.2f}'.format(av_price)
total_rev='${:,.2f}'.format(total_rev)
df_PAnalysis=pd.DataFrame({"Number of Unique Items": [item_ID], 
                          "Average Price":[av_price], 
                          "Number of Purchases":[num_purchase], 
                          "Total Revenue":[total_rev]})
df_PAnalysis.head()
```

```python
clean_purchase_data = purchase_data.drop_duplicates("SN")
gender = clean_purchase_data["Gender"].value_counts()
gender_percentage = gender / Player_info *100


df_gend=pd.DataFrame({"Total Count":gender, 
                     "Percentage of Players":gender_percentage.astype(float).map("{:,.2f}%".format)})
df_gend.head()
```

```python
av_price=purchase_data.groupby(["Gender"]).Price.mean()
gender_org = purchase_data["Gender"].value_counts()
total_rev=purchase_data.groupby(["Gender"]).Price.sum()
av_price_clean=total_rev/gender


df_Gender_Analysis=pd.DataFrame({"Purchase Count":gender_org, 
                          "Average Purchase Price":av_price.astype(float).map("${:,.2f}".format), 
                          "Total Purchase Value":total_rev.astype(float).map("${:,.2f}".format), 
                          "Avg Total Purchase per Person":av_price_clean.astype(float).map("${:,.2f}".format)})

df_Gender_Analysis.head()
```

```python
bins=[0,9,14,19,24,29,34,39,100]
group_ages= ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
clean_purchase_data["Total Ages"]=pd.cut(clean_purchase_data["Age"], bins, labels=group_ages)
totalages=clean_purchase_data["Total Ages"].value_counts()
purchase_data["Total Ages"]=pd.cut(purchase_data["Age"], bins, labels=group_ages)

totalper=totalages/Player_info * 100
df_totalages = pd.DataFrame({"Total Count":totalages, 
                            "Percentage of Players":totalper.astype(float).map("{:,.2f}%".format)})
df_totalages.sort_index(0)
# df_totalages.head(10)
```

```python
av_price_age = purchase_data.groupby(["Total Ages"]).Price.mean()
total_org_age =purchase_data["Total Ages"].value_counts()
total_rev_age=purchase_data.groupby(["Total Ages"]).Price.sum()
totalagestotal = clean_purchase_data["Total Ages"].value_counts()
av_price_clean_age=total_rev_age/totalagestotal

df_Total_Analysis_age=pd.DataFrame({"Purchases Count":total_org_age, 
                          "Average Price":av_price_age.astype(float).map("${:,.2f}".format), 
                          "Total Purchase Value":total_rev_age.astype(float).map("${:,.2f}".format), 
                          "Avg Total Purchase per Person":av_price_clean_age.astype(float).map("${:,.2f}".format)})

df_Total_Analysis_age.head(8)
```

```python
total_player_purchase  = purchase_data.groupby(["SN"]).Price.sum()
totave_player_price = purchase_data.groupby(["SN"]).Price.mean()
tot_purchase_value = purchase_data.groupby(["SN"]).Price.count()
purchase_data["Total Ages"]=pd.cut(purchase_data["Age"], bins, labels=group_ages)

df_total_SN=pd.DataFrame({"Purchase Count":tot_purchase_value, 
                          "Average Purchase Price":totave_player_price.astype(float).map("${:,.2f}".format), 
                          "Total Purchase Value":total_player_purchase})
df_total_SN=df_total_SN.sort_values(["Total Purchase Value"], ascending=False)
df_total_SN["Total Purchase Value"]=df_total_SN["Total Purchase Value"].map("${:,.2f}".format)
df_total_SN.head()
```
